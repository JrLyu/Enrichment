## Model Set Up

### Sigmoid Function

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
```

### Logistic Regression

```python
def z(x, w, b):
    z_wb = np.dot(x, w) + b
    mu = np.mean(z_wb)
    sgm = np.std(z_wb)
    z = (z_wb - mu) / sgm
    return z
```

```python
w_init = np.array([1,1,1])
b_init = 0

z_init = z(x, w_init,b_init)
print(z_init)
```

### Gradient Descent

```python
def compute_gradient_logistic(x, y, w, b): 
    m,n = x.shape
    dj_dw = np.zeros((n,))                          
    dj_db = 0.
    f_wb = sigmoid(z_score_normal(z(x, w, b)))

    for i in range(m):
        f_wb_i = f_wb[i]     
        err_i  = f_wb_i  - y[i]                      
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * x[i,j]     
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m                                  
    dj_db = dj_db/m                                  
        
    return dj_dw, dj_db
```

```python
dj_dw_init, dj_db_init = compute_gradient_logistic(x, y, w_init, b_init)
print(f"dj_db: {dj_db_init}" )
print(f"dj_dw: {dj_dw_init.tolist()}" )
```

```python
def gradient_descent(x, y, w_inint, b_inint, alpha, num_iters): 
    J_history = []
    w = copy.deepcopy(w_init)
    b = b_init
    
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient_logistic(x, y, w, b)   
        
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db               
      
        if i<100000:   
            J_history.append( compute_cost_logistic(x, y, w, b) )
            
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
        
    return w, b, J_history
```

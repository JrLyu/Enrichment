# Function Drawing

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

Linear Function

```python
x = np.arange(0, 10, 0.1)
y = x * 2

plt.title("Linear Function")
plt.plot(x,y)

plt.show()
```

Quadratic Function

```python
x = np.arange(-10, 10, 0.1)
y = x ** 2

plt.title("Quadratic Function")
plt.plot(x,y)

plt.show()
```

Exponential Function

```python
x = np.arange(-10,10,0.1)
y = 2 ** x

plt.title("Exponential Function")
plt.plot(x,y)

plt.show()
```

Natural Exponential Function

```python
e = math.e
x = np.arange(-10,10,0.1)
y = e ** x

plt.title("Natural Exponential Function")
plt.plot(x,y)

plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

plt.title("Sine Function")
plt.plot(x,y)

plt.show()
```

3D Curves and Scatter Plots

```python
# 3D Curves
fig1 = plt.figure()
ax1 = plt.axes(projection="3d")

z = np.linspace(0,13,1000)
x = 5 * np.sin(z)
y = 5 * np.cos(z)

# Scatter Plot
zd = 13 * np.random.random(100)
xd = 5 * np.sin(zd)
yd = 5 * np.cos(zd)

ax1.scatter3D(xd,yd,zd, cmap='Blues')
ax1.plot3D(x,y,z,'gray')
plt.show()
```

3D Surfaces and Contour

```python
fig3 = plt.figure()
ax3 = plt.axes(projection="3d")

xx = np.arange(-5,5,0.5)
yy = np.arange(-5,5,0.5)
X, Y = np.meshgrid(xx, yy)
Z = np.sin(X) + np.cos(Y)

ax3.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'rainbow')
ax3.contour(X, Y, Z, zdim='z', offset=-2, cmap='rainbow')
plt.show()
```

Plot contour on different surfaces

```python
fig4 = plt.figure()
ax4 = plt.axes(projection='3d')

xx = np.arange(-5,5,0.1)
yy = np.arange(-5,5,0.1)
X, Y = np.meshgrid(xx, yy)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax4.plot_surface(X, Y, Z, alpha = 0.3, cmap = "winter")
ax4.contour(X, Y, Z, zdir = "z", offset = -3, cmap = "rainbow")
ax4.contour(X, Y, Z, zdir = "x", offset = -6, cmap = "rainbow")
ax4.contourf(X, Y, Z, zdir = "y", offset = 6, cmap = "rainbow") # Fill the contour

ax4.set_xlabel("X")
ax4.set_xlim(-6,4)
ax4.set_ylabel("Y")
ax4.set_ylim(-4,6)
ax4.set_zlabel("Z")
ax4.set_zlim(-3,3)

plt.show()
```

Example. 1
Draw the multivariable function 
$$
z=f(x,y)=\sqrt{x^2+y^2-4}
$$

```python
fig5 = plt.figure()
ax5 = plt.axes(projection="3d")

xx = np.arange(-10,10,0.1)
yy = np.arange(-10,10,0.1)
X, Y = np.meshgrid(xx, yy)
Z = np.sqrt(X**2 + Y**2 - 4)

ax5.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = "winter")
plt.show()
```

Example.2 Draw the following vector function.
$$
\vec{r}(t)=\Big\langle f(t),\ g(t),\ h(t)\Big\rangle=\Big\langle \ln{(t+1)},\ \frac{t}{\sqrt{9-t^2}},\ 2^t\Big\rangle
$$

```python
fig5 = plt.figure()
ax5 = plt.axes(projection="3d")
    
t = np.arange(-0.9,3,0.1)

x = np.log(t+1)
y = t / np.sqrt(9 - t **2)
z = 2 ** t
    
ax5.plot3D(x, y, z, "black")
ax5.set_xlabel("X")
ax5.set_ylabel("Y")
ax5.set_zlabel("Z")
ax5.set_xlim(np.log(0.1),np.log(3.9))
ax5.set_ylim(-0.9 / np.sqrt(9 - -0.9 **2),2.9 / np.sqrt(9 - 2.9 **2))
ax5.set_zlim(2**-0.9,2**2.9)
plt.show()
```
